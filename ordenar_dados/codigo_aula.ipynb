```python
import pandas as pd

pedidos = pd.read_csv('./pedidos.csv')
pedidos.head(5)

# Ordenação crescente com base em coluna
pedidos_1 = pedidos.sort_values(by='preco_unitario')
pedidos_1.head(5)
# Ordenação decrescente com base em coluna
pedidos_2 = pedidos.sort_values(by='preco_unitario',ascending=False)
pedidos_2.head(5)
# Ordenação por Múltiplas Colunas
pedidos_3 = pedidos.sort_values(by=['categoria','quantidade'])
pedidos_3.head(5)
# Ordenação por data
pedidos['data_pedido'] = pd.to_datetime(pedidos['data_pedido'])
pedidos_4 = pedidos.sort_values(by='data_pedido',ascending=False)
pedidos_4.head(5)
```
