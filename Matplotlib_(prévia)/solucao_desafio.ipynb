```python
# Baixe o arquivo visitas_parque.csv e faça os seguintes desafios
# Monte e exiba um gráfico de barra(bar) com estes dados
# Não esqueça de colocar uma legenda textual, representando cada dia da semana
import pandas as pd
import matplotlib.pyplot as plt

# obtenha a média de visitas agrupadas por dias da semana
visitas = pd.read_csv('./visitas_parque.csv')
visitas['Data'] = pd.to_datetime(visitas['Data'])
visitas_por_estacao = visitas.groupby(visitas['Data'].dt.dayofweek)['Visitas'].sum()

visitas_por_estacao.plot(kind='bar') # bar, line, pie
plt.title('Total de visitas por dia da semana')
plt.xlabel('Dia')
plt.ylabel('Total de Visitas')
plt.xticks(range(7), ['Seg', 'Ter', 'Qua', 'Qui','Sex','Sab','Dom'])
plt.show()

```
