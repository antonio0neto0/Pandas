```python
import pandas as pd
import matplotlib.pyplot as plt

# Exiba um gráfico de barra que mostrar o total de vendas por mês
# Para gerar gráficos simples, precisamos primeiro de 2 pontos de dados(X e Y)
vendas = pd.read_csv('./vendas.csv')
vendas['Data'] = pd.to_datetime(vendas['Data'])
# Agrupar o total de vendas por mês
vendas_por_mes = vendas.groupby(vendas['Data'].dt.month)['Valor'].sum()

# Como criar um gráfico de barra simples
vendas_por_mes.plot(kind='bar') # bar, pie, line
plt.title('Vendas Totais por Mês')
plt.xlabel('Mês')
plt.ylabel('Valor Total de Vendas')
plt.xticks(range(12),['Jan','Fev','Mar','Abr','Mai','Jun','Jul','Ago','Set','Out','Nov','Dez'])
plt.show()
```
