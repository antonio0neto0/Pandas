# Desafio 1

```python
import pandas as pd

mercado = pd.read_csv('./vendas-supermercado.csv')
# Desafio 1
# Lucro Médio por Subcategoria e Segmento
# Desafio: Calcule o lucro médio por subcategoria e segmento
# Opção 1 
lucro_medio = mercado.groupby(['Subcategoria', 'Segmento'])['Lucro'].mean()
# Opção 2
lucro_medio_2 = mercado.groupby(['Subcategoria','Segmento']).agg({'Lucro':'mean'})
lucro_medio_2




```

# Desafio 2

```python
import pandas as pd

mercado = pd.read_csv('./vendas-supermercado.csv')

# Desafio 2
# Quantidade Total e Desconto Médio por Cidade
# Desafio: Determine a quantidade total de produtos vendidos e o desconto médio aplicado em cada cidade usando a função "agg"
cidade_resumo = mercado.groupby('Cidade').agg({'Quantidade': 'sum', 'Desconto': 'mean'})
cidade_resumo.head(5)





```

# Desafio 3

```python
import pandas as pd

mercado = pd.read_csv('./vendas-supermercado.csv')

# Desafio 3
# Análise Detalhada por Cliente
# Desafio: Crie um resumo detalhado das vendas por cliente, incluindo o total de vendas, o número de pedidos e o lucro médio usando a função "agg"
cliente_resumo = mercado.groupby('Nome do Cliente').agg({'Vendas': 'sum', 'ID do Pedido': 'count', 'Lucro': 'mean'})
cliente_resumo



```
