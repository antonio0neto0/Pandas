```python
import pandas as pd

pedidos = pd.read_csv('pedidos.csv')

pedidos.head(5)
#ver os tipops das colunas
pedidos.dtypes
#converter para data
pedidos['data_pedido'] = pd.to_datetime(pedidos['data_pedido'])
pedidos.dtypes

#filtrar por data exata
pedidos_2022_05_13 = pedidos[pedidos['data_pedido'] == '2022-05-13']
pedidos_2022_05_13.head(5)

#filtrar apos ou antes de alguma data
pedidos_apos_2023 = pedidos[pedidos['data_pedido'] >= '2023-01-01']
pedidos_apos_2023.head(5)

#Entre 2 datas
pedidos_meses_1_2_3 = pedidos[(pedidos['data_pedido'] >= '2022-01-01') & (pedidos['data_pedido'] <= '2022-03-31')]
pedidos_meses_1_2_3

#filtra por mes ou ano
pedidos_2022 = pedidos[pedidos['data_pedido'].dt.year == 2022]

pedidos_mes3 = pedidos[pedidos['data_pedido'].dt.month == 3]

pedidos_teste = pedidos[(pedidos['data_pedido'].dt.year == 2022) & (pedidos['data_pedido'].dt.month == 3)]
pedidos_teste
```
