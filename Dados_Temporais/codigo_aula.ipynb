```python
import pandas as pd

supermercado = pd.read_csv('./vendas-supermercado.csv')
# Verificar tipos
supermercado.dtypes
# Converter para datatime
supermercado['Data do Pedido'] = pd.to_datetime(supermercado['Data do Pedido'])
supermercado.dtypes
# Extrair Ano, Mês e dia
supermercado['ano'] = supermercado['Data do Pedido'].dt.year
supermercado['mes'] = supermercado['Data do Pedido'].dt.month
supermercado['dia'] = supermercado['Data do Pedido'].dt.day
supermercado.head(5)
# Agrupar por alguma coluna + data 
# Mostrar o total de vendas, agrupados por ano
vendas_por_ano = supermercado.groupby('mes')['Vendas'].mean()
vendas_por_ano

supermercado.head(5)
```
