# Tarefa 1
Analise o desempenho das diferentes categorias de produtos em termos de média de vendas por categoria, vendas totais, máxima e mínima de cada categoria.

```python

import pandas as pd

# Carregar o arquivo 'dados_ecommerce.csv'
ecommerce = pd.read_csv('dados_ecommerce.csv')

# Agrupar os dados por categoria e calcular a soma total das vendas para cada categoria
vendas_por_categoria = ecommerce.groupby('Categoria')['Vendas'].agg(['mean','sum','max','min'])
# Exibir os resultados
vendas_por_categoria
```


# Tarefa 2

O que fazer: Calcule o total de vendas para cada mês e demonstre gráficamente(através de um gráfico de barra simples)

```python
import pandas as pd
import matplotlib.pyplot as plt

ecommerce = pd.read_csv('dados_ecommerce.csv')
ecommerce['Data do Pedido'] = pd.to_datetime(ecommerce['Data do Pedido'])
ecommerce['Mês'] = ecommerce['Data do Pedido'].dt.month
vendas_por_mes = ecommerce.groupby('Mês')['Vendas'].sum()

# Plotando as vendas por mês em um gráfico de barras
vendas_por_mes.plot(kind='bar')
plt.title('Vendas Totais por Mês')
plt.xlabel('Mês')
plt.ylabel('Vendas Totais')
plt.xticks(range(12), ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'])
plt.show()
```

# Tarefa 3

O que fazer: Identifique a total de vendas em dias úteis(dias de semana) e o total de vendas em finais de semana e exiba esses resultados na tela.


```python

import pandas as pd
import matplotlib.pyplot as plt

# Carregar o arquivo 'dados_ecommerce.csv'
ecommerce = pd.read_csv('dados_ecommerce.csv')

# Converter a coluna 'Data do Pedido' para datetime e extrair o dia da semana
ecommerce['Data do Pedido'] = pd.to_datetime(ecommerce['Data do Pedido'])
ecommerce['Dia da Semana'] = ecommerce['Data do Pedido'].dt.dayofweek

# Definir uma função para identificar finais de semana
def e_final_de_semana(data_pedido):
    dia_da_semana = data_pedido.dayofweek
    return 1 if dia_da_semana >= 5 else 0

# Aplicar a função para criar uma coluna 'Final de Semana'
ecommerce['Final de Semana'] = ecommerce['Data do Pedido'].apply(e_final_de_semana)

# Calcular a média de vendas nos dias úteis e nos finais de semana
vendas_dia_semana = ecommerce[ecommerce['Final de Semana'] == 0].groupby('Dia da Semana')['Vendas'].sum()
vendas_final_semana = ecommerce[ecommerce['Final de Semana'] == 1].groupby('Dia da Semana')['Vendas'].sum()

# Exibir os resultados
print("Média de Vendas em Dias Úteis: ", vendas_dia_semana)
print("Média de Vendas em Finais de Semana: ", vendas_final_semana)

# Plotando as vendas por mês em um gráfico de barras
vendas_dia_semana.plot(kind='bar')
plt.title('Vendas Dias de Semana')
plt.xlabel('Mês')
plt.ylabel('Vendas Totais')
plt.xticks(range(5),['segunda','terça','quarta','quinta','sexta'])
plt.show()

vendas_final_semana.plot(kind='bar')
plt.title('Vendas Finais de Semana')
plt.xlabel('Mês')
plt.ylabel('Vendas Totais')
plt.xticks(range(2),['sábado','domingo'])
plt.show()
```
