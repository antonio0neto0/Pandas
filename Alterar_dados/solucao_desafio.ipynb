```python
import pandas as pd
# Desafios Práticos:
pedidos = pd.read_csv('./pedidos.csv')

# Desafio 1: Alterar o valor da terceira linha da coluna 'Produto' de 'Relógio' para 'Produto 3'.
pedidos.at[2, 'produto'] = 'Produto 3'

# Desafio 2: Substituir todos os métodos de pagamento 'Boleto' por 'À Vista'
pedidos.loc[pedidos['metodo_pagamento'] == 'Boleto', 'metodo_pagamento'] = 'À Vista'

# Desafio 3: Reduzir o custo de envio em 5% para todos os pedidos
def reduzir_custo(custo):
    return custo * 0.95
pedidos['custo_envio'] = pedidos['custo_envio'].apply(reduzir_custo)
pedidos.head()
```
