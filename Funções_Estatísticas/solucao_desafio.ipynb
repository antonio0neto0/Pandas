```python
import pandas as pd
import numpy as np


pedidos = pd.read_csv('pedidos.csv')
# Desafio 1: Média e Mediana do Custo de Envio
media_custo_envio = pedidos['custo_envio'].mean()
mediana_custo_envio = pedidos['custo_envio'].median()
print(f'Média do custo de envio: {media_custo_envio}')
print(f'Mediana do custo de envio: {mediana_custo_envio}')

# Desafio 2: Desvio Padrão das Quantidades
desvio_padrao_quantidade = pedidos['quantidade'].std()
print(f'Desvio padrão da quantidade de itens por pedido: {desvio_padrao_quantidade}')

# Desafio 3: Amplitude do custo de envio
amplitude_custo_envio = pedidos['custo_envio'].max() - pedidos['custo_envio'].min()
print(f'Amplitude do preço unitário: {amplitude_custo_envio}')
```
